```python
import pandas as pd #นำเข้า pandas แทนที่ด้วย pd
import os
```

```python
from google.colab import drive #นำเข้าข้อมูลจาก drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/dataviz_2021_data' #ไดร์ฟชื่อว่า dataviz_2021_data
```

```python
data_file_path = os.path.join(path,'pm-20-jan-2021.csv') #ไฟล์ pm-20-jan-2021.csv
data_df = pd.read_csv(data_file_path) #นามสกุลไฟล์ csv
```

```python
set(data_df['risk']) 
```

```python
data_df
```

```python
data1 = data_df[data_df['risk']== 'สถานบันเทิง']
data1
```

```python
data2 = data_df[data_df['risk']== 'สนามมวย']
data2
```

```python
data3 = data_df[data_df['risk']== 'สัมผัสผู้ป่วยยืนยัน']
data3
```

```python
 data4 = pd.concat([data1,data2])
 data5 = pd.concat([data4,data3])
 data5
```

```python
this_data = data5[['sex','age','risk']]
```

```python
this_data['sex'].replace({'ชาย':1,'หญิง':50}, inplace=True)
```

```python
this_data['risk'].replace({'สถานบันเทิง':'entertainment place','สนามมวย':'boxing stadium','สัมผัสผู้ป่วยยืนยัน':'Close Contact'}, inplace=True)
```

```python
this_data
```

```python
pd.plotting.parallel_coordinates(this_data,'risk',colormap='Set1' )
```
